```python
import pandas as pd

# Read csv files.
df1 = pd.read_csv('RegularSeasonDetailedResults.csv')
df2 = pd.read_csv('TourneyDetailedResults.csv')

# Merge data frames and sort by year and day number.

df = pd.concat([df1,df2], ignore_index=True)
df.sort_values(by = ['Season','Daynum'], ascending=True)
```

```python
df_stats = df.iloc[:,6:]
df_stats[:5]
```

```python
stat_cols = df_stats.columns.tolist()
stat_cols
```

```python
import itertools
tee = itertools.product(stat_cols, stat_cols)
#print(tee)
bar = list() 
for a, b in tee:
    #print(a, b)
    if(a != b) and (a[1:] == b[1:]):    
        if(a[0] == 'W'):
            #print(a, b)
            bar.append((a, b))
    #break
print(bar)
```

```python
new_cols = list(itertools.chain(*bar))
print(new_cols)
print(len(new_cols))
```

```python
df_stats = df_stats[new_cols]
df_stats[:5]
```

```python
df_diff = pd.DataFrame()
for e in bar:
    #print(e[0][1:]+'_diff')
    #print((df_stats[e[0]] - df_stats[e[1]])/df_stats[e[0]])
    col_name = e[0][1:]+'_diff'
    #if (df_stats[e[0] != 0]):
    df_diff[col_name] = (df_stats[e[0]] - df_stats[e[1]])/(df_stats[e[0]]+1.0)
    #else:
    
    #break
df_diff
```

```python
df_new = pd.concat([df.iloc[:,:6], df_diff], axis = 1)
df_new[:5]
```

```python
df_new.shape
```

```python
# Create list of teams.

df_teams = pd.read_csv('Teams.csv')
teams = df_teams['Team_Id'].tolist()

# Create list of attributes.

attributes = list(df_new)[6:]
```

```python
attributes[:5]
```

```python
# Create data structure.

foo = {}

for team in teams:
    foo[team] = {}
    for attribute in attributes:
        foo[team][attribute] = []
```

```python
import numpy as np

# Update.
def update_attribute(foo, team, attribute, value):
    if len(foo[team][attribute]) < 3:
        foo[team][attribute].append(value)
    else:
        dummy = foo[team][attribute].pop(0)
        foo[team][attribute].append(value)


# Calculate.
def calculate_trailing(foo, team, attribute):
    if len(foo[team][attribute]) < 3:
        return 0
    else:
        return np.mean(foo[team][attribute])
```

```python
years = range(2003,2018)

# 3 day trailing average.
for year in years:
    for i in range(len(df.index)):
        if df_new.iloc[i]['Season'] == year:
            for attribute in attributes:
                if attribute[:1] == 'W':
                    update_attribute(foo, df.iloc[i]['Wteam'], attribute, df_new.iloc[i][attribute])
                else:
                    update_attribute(foo, df.iloc[i]['Lteam'], attribute, df_new.iloc[i][attribute])
            for attribute in attributes:
                if attribute[:1] == 'W':
                    df_new.set_value(i, attribute, calculate_trailing(foo, df_new.iloc[i]['Wteam'], attribute))
                else:
                    df_new.set_value(i, attribute, calculate_trailing(foo, df_new.iloc[i]['Lteam'], attribute))
```
